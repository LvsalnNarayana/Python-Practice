# Variables

- Constant and variable names should contain a combination of lowercase (a-z) or capital (A-Z) characters, numbers (0-9), or an underscore ( ).
- If you want to use underscore to separate two words in a variable name, do so.
- To declare a constant, use all capital letters as feasible.
- Never use special characters such as !, @, #, $, %, and so on.
- A variable/constant name should never begin with a number.



```python

AGE = 18

name = "John"

print(AGE)

print(name)
```
